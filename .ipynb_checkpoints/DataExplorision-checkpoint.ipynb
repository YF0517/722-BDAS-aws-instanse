```python
# Must be included at the beginning of each new notebook. Remember to change the app name.
import findspark
findspark.init('/home/ubuntu/spark-3.2.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('basics').getOrCreate()
```

```python
df = spark.read.csv('data/tmdb_5000_movies_fixed1_fix.csv')
```

```python
df.show()
```

```python
# Optionally, rename columns for better clarity
df = df.withColumnRenamed("_c0", "budget") \
       .withColumnRenamed("_c1", "id") \
       .withColumnRenamed("_c2", "original_language") \
       .withColumnRenamed("_c3", "popularity") \
       .withColumnRenamed("_c4", "revenue") \
       .withColumnRenamed("_c5", "runtime") \
       .withColumnRenamed("_c6", "status") \
       .withColumnRenamed("_c7", "title") \
       .withColumnRenamed("_c8", "vote_average") \
       .withColumnRenamed("_c9", "vote_count") \
       .withColumnRenamed("_c10", "genre") \
       .withColumnRenamed("_c11", "release_date") \
       .withColumnRenamed("_c12", "production_country")
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import col, to_date
from pyspark.sql.types import IntegerType, DoubleType, DateType

# Convert columns to the desired data types
df = df.withColumn("budget", df["budget"].cast(IntegerType()))
df = df.withColumn("id", df["id"].cast(IntegerType()))
df = df.withColumn("popularity", df["popularity"].cast(DoubleType()))
df = df.withColumn("revenue", df["revenue"].cast(IntegerType()))
df = df.withColumn("runtime", df["runtime"].cast(IntegerType()))
df = df.withColumn("vote_average", df["vote_average"].cast(DoubleType()))
df = df.withColumn("vote_count", df["vote_count"].cast(IntegerType()))
df = df.withColumn("release_date", to_date(df["release_date"], "yyyy-MM-dd"))

# Print the updated schema
df.printSchema()
# Get the number of rows in the DataFrame
num_rows = df.count()

# Print the number of rows
print(f"Number of rows in the DataFrame: {num_rows}")
```

```python
from itertools import combinations
from pyspark.sql.functions import corr

# Define the selected columns
selected_columns = ["budget", "revenue", "runtime", "vote_count", "popularity", "vote_average"]

# Compute the correlation matrix for each pair of columns
correlation_matrix = []

for col1, col2 in combinations(selected_columns, 2):
    corr_value = df.select(corr(col1, col2)).collect()[0][0]
    correlation_matrix.append((col1, col2, corr_value))

# Print the correlation matrix
print("Correlation Matrix:")
for col1, col2, corr_value in correlation_matrix:
    print(f"Correlation between {col1} and {col2}: {corr_value}")
```

### show each feature statistic

```python
# Get the list of column names
columns = df.columns

# Iterate over each column
for column in columns:
    print(f"Statistics for column '{column}':")
    
    # Calculate and display statistics for the current column
    df.describe(column).show()
```

### plot genre and production country

```python
# Group by the correct column name "_c10" (assuming it contains genre information)
genre_counts = df.groupBy("genre").count()

# Convert to Pandas DataFrame for plotting
genre_counts_pd = genre_counts.toPandas()

# Plotting the bar plot
import matplotlib.pyplot as plt

genre_counts_pd.set_index("genre", inplace=True)
genre_counts_pd.plot(kind='bar', legend=False)
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Genres')
plt.show()

# Show unique genres
unique_genres = df.select("genre").distinct().collect()
unique_genres = [row['genre'] for row in unique_genres]
print(unique_genres)
```

```python
import matplotlib.pyplot as plt

# Define the number of top countries to display
top_n = 10

# Convert to Pandas DataFrame for plotting
country_counts_pd = df.groupBy("production_country").count().toPandas()

# Sort and select the top N countries
top_countries_pd = country_counts_pd.sort_values('count', ascending=False).head(top_n)

# Plotting the bar plot for top N countries
top_countries_pd.plot(kind='bar', x='production_country', y='count', legend=False)
plt.xlabel('Production Country')
plt.ylabel('Frequency')
plt.title(f'Top {top_n} Production Countries by Frequency')
plt.show()

# Show unique genres
unique_country = df.select("production_country").distinct().collect()
unique_country = [row['production_country'] for row in unique_country]
print(unique_country)
```

### plot budget and revenue

```python
from scipy.interpolate import make_interp_spline
import numpy as np

# Convert Spark DataFrame to Pandas DataFrame
df_pandas = df.toPandas()

# Drop rows with missing values and duplicates
df_clean = df_pandas.copy().dropna().drop_duplicates(subset=['budget'])

# Sort values by 'budget'
df_clean = df_clean.sort_values(by='budget')

# Create a spline interpolation function
x = df_clean['budget']
y = df_clean['revenue']
x_smooth = np.linspace(x.min(), x.max(), 500)  
spl = make_interp_spline(x, y, k=1)  # Adjust the smoothness (degree of the spline) as needed
y_smooth = spl(x_smooth)

# Plot with the smoothed line
plt.plot(x_smooth, y_smooth, label='Revenue vs Budget (Smoothed)')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Revenue vs Budget')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Convert PySpark DataFrame to Pandas DataFrame
pandas_df = df.select("vote_average", "revenue").toPandas()

# Plot the scatter plot
plt.plot(pandas_df["vote_average"], pandas_df["revenue"], 'o', label='Revenue vs vote_average')

plt.xlabel('vote_average')  # Adjust the x-axis label if needed
plt.ylabel('Revenue')  # Adjust the y-axis label if needed
plt.title('Revenue vs vote_average')  # Adjust the title if needed
plt.legend()  # Show legend
plt.grid(True)  # Add gridlines
plt.show()
```

```python
# Filter out rows with missing or NaN values in the revenue column
df_clean = df.filter(df['revenue'].isNotNull())

# Aggregate revenue by genre
genre_revenue = df_clean.groupBy('genre').sum('revenue')

# Convert PySpark DataFrame to Pandas DataFrame
pandas_genre_revenue = genre_revenue.toPandas()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(pandas_genre_revenue['sum(revenue)'], labels=pandas_genre_revenue['genre'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Revenue by Genre')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

```python


import matplotlib.pyplot as plt
# Aggregate revenue by genre
country_revenue = df_clean.groupBy('production_country').sum('revenue')

# Convert PySpark DataFrame to Pandas DataFrame
pandas_production_country_revenue = country_revenue.toPandas()

# Filter out rows with null or NaN values in the 'sum(revenue)' column
pandas_production_country_revenue = pandas_production_country_revenue.dropna(subset=['sum(revenue)'])

# Define the number of top countries to display
top_n = 5

# Sort by 'sum(revenue)' in descending order and select the top N production countries
top_revenue_countries = pandas_production_country_revenue.sort_values(by='sum(revenue)', ascending=False).head(top_n)

# Create pie chart for the top N production countries
plt.figure(figsize=(8, 8))
plt.pie(top_revenue_countries['sum(revenue)'], labels=top_revenue_countries['production_country'], autopct='%1.1f%%', startangle=140)
plt.title('Top {} Production Countries by Revenue'.format(top_n))
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

### checking null value
