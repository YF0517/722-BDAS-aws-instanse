```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
import numpy as np
# Assu ming 'df' is your DataFrame with 'budget' and 'revenue' columns
df_clean= df.dropna()
df_clean.drop_duplicates(subset=['budget'], inplace=True)
df_clean = df_clean.sort_values(by='budget')
# Create a spline interpolation function
x = df_clean['budget']
y = df_clean['revenue']
x_smooth = np.linspace(x.min(), x.max(), 500)  
# Adjust the smoothness
spl = make_interp_spline(x, y, k=1)  
# k is the degree of the spline; adjust as needed
y_smooth = spl(x_smooth)
# Plot with the smoothed line
plt.plot(x_smooth, y_smooth, label='Revenue vs Budget (Smoothed)')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Revenue vs Budget')
plt.grid(True)
plt.legend()
plt.show() 
```

```python
# Aggregate revenue by genre
genre_revenue = df.groupby('genre')['revenue'].sum()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(genre_revenue, labels=genre_revenue.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Revenue by Genre')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

```python
# Aggregate revenue by production country
production_country_revenue = df.groupby('production_country')['revenue'].sum()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(production_country_revenue, labels=production_country_revenue.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Revenue by production country')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```
